## MATRIZ DE CORRELAÇÃO

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>O <b>coeficiente de correlação</b> é uma medida de associação linear entre duas variáveis e situa-se entre <b>-1</b> e <b>+1</b> sendo que <b>-1</b> indica associação negativa perfeita e <b>+1</b> indica associação positiva perfeita.</p>



```python
f, ax = plt.subplots(figsize = (20, 20))
sns.heatmap(dados.corr(), annot = True)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> O COEFICIENTE DE CORRELAÇÃO é uma importante medida estatística na análise de um modelo de regressão, pois informa a dependência linear entre a variável explicada (y) e explicativa (X), sendo que:
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>|r| = 0 --> Correlação Nula
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>0 < |r| <= 0,30 --> Correlação Fraca
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>0,30 < |r| <= 0,70 --> Correlação Média
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>0,70 < |r| <= 0,90 --> Correlação Forte
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>0,90 < |r| <= 0,99 --> Correlação Fortíssima
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>|r| = 1 --> Correlação Perfeita
 
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> Visto isso, a correlação utilizada no modelo foi maior ou igual a Média, coeficiente > 0,30.
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> Sendo elas:

## APLICANDO A TRANSFORMAÇÃO LOGARÍTMICA AOS DADOS

```python
np.log(0)
```

```python
dados['log_price'] = np.log(dados['price'])
dados['log_bedrooms'] = np.log(dados['bedrooms'])
dados['log_bathrooms'] = np.log(dados['bathrooms'])
dados['log_sqft_living'] = np.log(dados['sqft_living'])
dados['log_view'] = np.log(dados['view']+1)
dados['log_grade'] = np.log(dados['grade'])
dados['log_sqft_above'] = np.log(dados['sqft_above'])
dados['log_sqft_basement'] = np.log(dados['sqft_basement']+1)
dados['log_sqft_living15'] = np.log(dados['sqft_living15'])
```

## DISTRIBUIÇÃO DE FREQUÊNCIAS DA VARIÁVEL *DEPENDENTE TRANSFORMADA* (y)

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> Anteriormente, vimos que não existia uma relação entre a variável dependente e as variáveis explicativas, pois não conseguimos identificar uma relação linear entre elas. Além disso, descobrimos que a variável dependente que estamos analisando é assimétrica à direita.
Após a transformação das variáveis, podemos observar que a variável dependente está mais assimétrico e é possível verificar que existem uma relação linear com as variáveis explicativas.

## REGRESSÃO LINEAR

```python
y = dados['log_price']
```

```python
X = dados[[ 'log_bedrooms', 'log_bathrooms', 'log_sqft_living', 'log_view', 'log_grade','log_sqft_above', 'log_sqft_basement', 'log_sqft_living15']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 1000)
```

```python
X_train_com_constante = sm.add_constant(X_train)
```

```python
X_train_com_constante
```

```python
modelo_statsmodels = sm.OLS(y_train, X_train_com_constante, hasconst = True).fit()
```

# <font color='blue' style='font-size: 30px;'>AVALIANDO O MODELO ESTIMADO</font>
<hr style='border: 2px solid blue;'>

```python
print(modelo_statsmodels.summary())
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> O quadrado do coeficiente de correlação resulta no coeficiente de determinação, que indica o poder de explicação do modelo, em função das variáveis independentes consideradas.
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'> Analisando os dados da tabela acima, o valor de R² foi de 58%


# <font color='blue' style='font-size: 30px;'>ESTIMANDO O MODELO COM OS DADOS DE TREINO</font>
<hr style='border: 2px solid blue;'>

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
print('R² = {}'.format(modelo.score(X_train, y_train).round(2)))
```

```python
y_previsto = modelo.predict(X_test)
```

```python
print('R² = %s' % metrics.r2_score(y_test, y_previsto).round(2))
```
